```python
from typing import List, Tuple
from Calendar.Calculation import Calculation
from pymeeus.Epoch import Epoch
from pymeeus.JupiterMoons import JupiterMoons

import matplotlib.pyplot as plt
```

```python
epoch_start = Epoch()
epoch_start.set(2020, 1, 1, 0)
epoch_end = Epoch()
epoch_end.set(2020, 3, 1, 0)
# 1 s in jd = 1.157401129603386e-05
time_step = 60 * 120 * 1.157401129603386e-05

calc = Calculation(epoch_start, epoch_end, time_step)
```

```python
plt.plot(calc.distances[0][0])
plt.grid(True)
plt.ylim([0, 1])

print(len(calc.timing_lists[0][1]))
print(calc.timing_lists[0][1])
```

```python
for row in range(len(calc.timing_lists)):
    for col in range(len(calc.timing_lists[row]) - 1):
        for timing in calc.timing_lists[row][col]:
            print("Sat: ", row + 1, ", ", col, ", ", timing[1])
            print(timing[2].get_full_date())
            print()

```
